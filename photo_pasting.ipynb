```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

addWeighted function blend two images by formula ;
new_pixel = alpha * pix_1 + beta * pix_2 + gamma

```python
from PIL import Image
dog_pic = Image.open('dog_backpack.jpg')
img1 = np.asarray(dog_pic)
plt.imshow(img1)
```

```python
img2 = np.asarray(cv2.imread('watermark_no_copy.png'))
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
new_img1 = img1.copy()
new_img2 = img2.copy()
```

```python
pic1 = cv2.resize(new_img1, (1000,1000))
pic2 = cv2.resize(new_img2, (1000,1000))
```

```python
plt.imshow(new_img1)
```

```python
blended = cv2.addWeighted(pic1, 0.5, pic2, 0.5, 10)
plt.imshow(blended)
```

```python
# Resize to Width=400, Height=200
new_pic2 = cv2.resize(new_img2, (400, 200))
new_pic1 = new_img1.copy()

# Slice to match: [0:height, 0:width] -> [0:200, 0:400]
new_pic1[0:200, 0:400] = new_pic2
plt.imshow(new_pic1)
```
