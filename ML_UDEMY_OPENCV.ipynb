```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
cv_img = cv2.imread('00-puppy.jpg')
```

```python
print(type(cv_img))
```

```python
print(cv_img.shape)
```

```python
plt.imshow(cv_img)
# in opencv --> opposite of RGB
```

```python
fix_img = cv2.cvtColor(cv_img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
# if we want to make it greyscale then from the upload we can take it as greyscale
img_grey = cv2.imread('00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
img_grey.shape
```

```python
plt.imshow(img_grey, cmap = "gray")
```

```python
new_size = cv2.resize(fix_img, (1000,500))
plt.imshow(new_size)
```

```python
h_ratio = 0.6
w_ratio = 0.5
```

```python
new_size2 = cv2.resize(fix_img, (0,0),fix_img, w_ratio, h_ratio)
print(new_size2.shape)
plt.imshow(new_size2)
```

```python
#  FLIP ABOUT X AXIS , FOR Y AXIS USE 1 , FOR BOTH AXIS FLIP USE -1
new_img3 = cv2.flip(fix_img, 0)
plt.imshow(new_img3)
```

```python
fig = plt.figure(figsize = (5,8))
#This creates a blank canvas (a "figure") to draw on. The figsize=(10,8)
#tells the computer to make the canvas 10 inches wide and 8 inches tall. It ensures your puppy photo looks nice and big!
ax = fig.add_subplot(111)
#What does 111 mean? It’s a shortcut grid code. It means: "Create a grid that is 1 row by 1 column, and put this plot in the 1st spot.
#" In plain English: it just tells the computer to let this single plot take up the entire canvas.
ax.imshow(fix_img)
#This actually takes your puppy image (fix_img) and paints it onto that drawing area so you can look at it.
```
