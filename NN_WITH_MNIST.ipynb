```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train.shape
```

```python
single_image =x_train[0]
single_image
```

```python
plt.imshow(single_image, cmap = 'gray')
```

```python
print(y_train)
print(y_train.shape)
```

```python
from keras.utils import to_categorical
##you are transitioning from binary classification (0 or 1) to Multi-class classification
 ##(e.g., classifying images into cats, dogs, or birds, or predicting handwritten digits from 0 to 9).
##to_categorical converts integer class labels into a One-Hot Encoded format.
```

```python
## 10 shows 10 category will be generated
y_cat_test= to_categorical(y_test,10)
y_cat_train= to_categorical(y_train,10)
```

```python
y_cat_train[0]
```

```python
x_train[0].max()
```

```python
#normalising the trainin g srt manually
x_train = x_train/x_train.max()
x_test = x_test/x_test.max()
x_train.max()
```

```python
scaled_image = x_train[0]
scaled_image
```

```python
plt.imshow(scaled_image, cmap = 'gray')
```

```python
x_train.shape
```

```python
x_train = x_train.reshape(60000,28,28,1)
x_test = x_test.reshape(10000,28,28,1)
```

```python
x_test.shape
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense, Conv2D, MaxPool2D,Flatten
```

```python
model = Sequential()

model.add(Conv2D(filters = 32, kernel_size = (4,4), input_shape = (28,28,1), activation = 'relu'))

model.add(MaxPool2D(pool_size = (2,2)))

model.add(Conv2D(filters = 64, kernel_size = (3,3), activation = 'relu'))

model.add(MaxPool2D(pool_size = (2,2)))

model.add(Flatten())
##a Dense layer cannot accept a 2D image directly.
model.add(Dense(128, activation = 'relu'))

model.add(Dense(10, activation = 'softmax'))

model.compile(loss = 'categorical_crossentropy', optimizer = 'rmsprop', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train, y_cat_train, epochs = 2, verbose = 1)
```

```python
model.metrics_names
```

```python
model.evaluate(x_test, y_cat_test)
```

```python
from sklearn.metrics import classification_report
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
# Find the index of the maximum value along the columns (axis=1)
predicted_classes = np.argmax(predictions, axis=1)
predicted_classes
```

```python
report = classification_report(y_test, predicted_classes)
print(report)
```
