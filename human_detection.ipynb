```python
from ultralytics import YOLO
from PIL import Image
import matplotlib.pyplot as plt
```

```python
model = YOLO('yolov8n.pt')  # 'n' = nano, fastest version

# Run detection on any image URL
results = model.predict(
    source='https://ultralytics.com/images/bus.jpg',
    conf=0.5,      # confidence threshold — same as professor's 0.6
    save=True,     # saves result image
    project='/content/runs'  # where to save
)
```

```python
print(results)
```

```python

result_img = results[0].plot()  # draws boxes on image
result_rgb = result_img[:, :, ::-1]  # BGR to RGB

plt.figure(figsize=(12, 8))
plt.imshow(result_rgb)
plt.axis('off')
plt.title('YOLOv8 Detection')
plt.show()
```
